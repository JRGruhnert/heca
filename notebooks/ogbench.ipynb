```python
import numpy as np
import gymnasium as gym
import ogbench
```

```python

id: str = "visual-scene-play-v0"
mode: str = "task"
ob_type: str = "pixels"  # states, pixels
width: int = 256
height: int = 256
terminate_at_goal: bool = True
success_timing: str = "post"  # pre, post
physics_timestep: float = 0.002
control_timestep: float = 0.05
visualize_info: bool = False  # True adds semi-transparent goal-state overlays → color artifacts

```

```python
env= ogbench.make_env_and_datasets(
                dataset_name=id,
                env_only=True,
                dataset_only=False,
                #ob_type=ob_type,
                mode=mode,
                #visualize_info=visualize_info,
                width=width,
                height=height,
            )
assert isinstance(env, gym.Env)

obs, info = env.reset() 
print(info.keys())
print(info["goal"])

action = env.action_space.sample()
print(action)
```

```python
import matplotlib.pyplot as plt
assert isinstance(env, gym.Env)
obs, reward, done, truncated, info = env.step(action)
frame = env.render()
assert isinstance(frame, np.ndarray)
plt.imshow(frame)
plt.axis("off")
plt.show()
```
